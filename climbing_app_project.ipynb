```python
# libraries we need
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
def mask(path='./climbing_app_project/kilter_board_setting.jpeg'):
    # Python program to transform an image using
    # threshold.
    # Image operation using thresholding
    img = cv2.imread(path)

    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    ret, thresh = cv2.threshold(gray, 0, 255,
                                cv2.THRESH_BINARY_INV +
                                cv2.THRESH_OTSU)
    
    # Noise removal using Morphological
    # closing operation
    kernel = np.ones((3, 3), np.uint8)
    closing = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE,
                                kernel, iterations = 2)

    # Background area using Dilation
    bg = cv2.dilate(closing, kernel, iterations = 1)

    # Finding foreground area
    dist_transform = cv2.distanceTransform(closing, cv2.DIST_L2, 0)
    ret, fg = cv2.threshold(dist_transform, 0.02
                            * dist_transform.max(), 255, 0)
    # save image
    cv2.imwrite('./climbing_app_project/foreground.jpeg', fg)
    
    # show image
    cv2.imshow('image', fg)
    if cv2.waitKey(0) & 0xFF == ord('q'): 
        cv2.destroyAllWindows()
```

```python
def contour_recognition(path='./climbing_app_project/foreground.jpeg'):
    # Python code to find the co-ordinates of
    # the contours detected in an image.
    # inputs: path to picture
    # outputs: picture with recognized shapes/contours
    
    # Reading image
    font = cv2.FONT_HERSHEY_COMPLEX
    img2 = cv2.imread(path, cv2.IMREAD_COLOR)

    # Reading same image in another 
    # variable and converting to gray scale.
    img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)

    # Converting image to a binary image
    # ( black and white only image).
    _, threshold = cv2.threshold(img, 110, 255, cv2.THRESH_BINARY_INV)

    # Detecting contours in image.
    contours, _= cv2.findContours(threshold, cv2.RETR_TREE,
                                   cv2.CHAIN_APPROX_SIMPLE)

    # Going through every contours found in the image.
    for cnt in contours :

        approx = cv2.approxPolyDP(cnt, 0.009 * cv2.arcLength(cnt, True), True)

        # draws boundary of contours.
        cv2.drawContours(img2, [approx], 0, (0, 0, 255), 5) 

        # Used to flatted the array containing
        # the co-ordinates of the vertices.
        n = approx.ravel() 

    # save image
    cv2.imwrite('./climbing_app_project/contours.jpeg', img2)
    
    # Showing the final image.
    cv2.imshow('Contour outlines recognized', img2) 

    # Exiting the window if 'q' is pressed on the keyboard.
    if cv2.waitKey(0) & 0xFF == ord('q'): 
        cv2.destroyAllWindows()
    return contours
```
